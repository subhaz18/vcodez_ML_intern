```python
#Q1
def unique_list(lst):
    return list(set(lst))

sample = [1,2,3,3,3,4,5]
print(unique_list(sample))
```

```python
#Q2
def count_vowels(string):
    vowels = "aeiouAEIOU"
    return sum(1 for ch in string if ch in vowels)

print(count_vowels("Hello World"))
```

```python
#Q3
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
print(is_prime(10))
```

```python
#Q4
def calculator(a, b, op):
    if op == '+': return a + b
    elif op == '-': return a - b
    elif op == '*': return a * b
    elif op == '/': return a / b if b != 0 else "Error: Division by zero"
    else: return "Invalid operator"

print(calculator(10, 5, '+'))
print(calculator(10, 5, '/'))
```
