```python
#Q8
rand_arr = np.random.randint(1, 100, 10)

print("Random Array:", rand_arr)
print("Mean:", np.mean(rand_arr))
print("Median:", np.median(rand_arr))
print("Max:", np.max(rand_arr))
print("Min:", np.min(rand_arr))
```
