```python
for i in teams:
  if i=='Super Kings':
    df1['team'].replace(i,'CSK',inplace=True)
  elif i=='Punjab Kings':
    df1['team'].replace(i,'PBKS',inplace=True)
  elif i=='Capitals':
    df1['team'].replace(i,'DC',inplace=True)
  elif i=='Mumbai':
    df1['team'].replace(i,'MI',inplace=True)
  elif i=='Sunrisers':
    df1['team'].replace(i,'SRH',inplace=True)
  elif i=='Royals':
    df1['team'].replace(i,'RR',inplace=True)
  elif i=='Titans':
    df1['team'].replace(i,'GT',inplace=True)
  elif i=='Super Giants':
    df1['team'].replace(i,'LSG',inplace=True)
```

```python
# Create a mapping of full team names to their abbreviations
team_map = {
    'Super Kings': 'CSK',
    'KKR': 'KKR',
    'Titans': 'GT',
    'Royals': 'RR',
    'Capitals': 'DC',
    'Super Giants': 'LSG',
    'Punjab Kings': 'PBKS',
    'Sunrisers': 'SRH',
    'Mumbai': 'MI',
    'KKR': 'KKR',
    'RCB': 'RCB'
}

# Replace long team names in the 'match' column using the mapping
def abbreviate_match_name(match):
    for full, short in team_map.items():
        match = match.replace(full, short)
    return match.upper()

# Apply the function to the column
df1['match'] = df1['match'].apply(abbreviate_match_name)

# View the updated column
df1[['match']].head()
```

```python
df1
```

```python
import re

# Insert space before capital letters (except the first one)
df1['batsman'] = df1['batsman'].apply(lambda x: re.sub(r'(?<=[a-z])(?=[A-Z])', ' ', x))

# Optional: strip spaces in case of any leading/trailing whitespace
df1['batsman'] = df1['batsman'].str.strip()

# Preview the fixed names
df1[['batsman']].head(10)
```

```python
import pandas as pd

# Step 1: Mark "out" as 1 and "not_out" as 0
df1['is_out'] = df1['out_status'].apply(lambda x: 1 if x == 'O' or x == 'out' else 0)

# Step 2: Group by batsman and calculate total runs and number of outs
stats = df1.groupby('batsman').agg(
    total_runs=('runs', 'sum'),
    outs=('is_out', 'sum'),
    innings=('matchID', 'count')
).reset_index()

# Step 3: Calculate average (avoid division by zero)
stats['average'] = stats.apply(
    lambda row: row['total_runs'] / row['outs'] if row['outs'] > 0 else row['total_runs'],
    axis=1
)

# Step 4: Sort by highest average
stats = stats.sort_values(by='average', ascending=False)

# Display results
stats.head(10)
```
